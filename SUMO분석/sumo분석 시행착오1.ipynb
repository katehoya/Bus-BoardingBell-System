```python
target_edge = "760739036"
target_position = (1024.27,387.3)
search_radius = 50

def calculate_density():
    #SUMO-GUI 실행
    traci.start([sumo_binary, "-c", sumo_config])
    print("Simulation started!")

    time = 0
    density_data = {}

    try:
        while time < 10800:
            traci.simulationStep() 

            #특정 도로 내 차량 수
            vehicles_on_edge = traci.edge.getLastStepVehicleIDs(target_edge)
            edge_density = len(vehicles_on_edge)

            #특정 위치 반경 내 차량 수
            all_vehicles = traci.vehicle.getIDList()
            position_density = 0
            for vehicle_id in all_vehicles:
                x, y = traci.vehicle.getPosition(vehicle_id)
                distance = ((x - target_position[0])**2 + (y - target_position[1])**2)**0.5
                if distance <= search_radius:
                    position_density += 1

            #결과 저장
            density_data[time] = {
                "edge_density" : edge_density,
                "position_density": position_density,
            }

            print(f"Time: {time}s, Edge Density: {edge_density}, Position Density: {position_density}")
            time +=1

    finally:
        traci.close()
        print("Simulation ended")

    return density_data

if __name__ == "__main__":
    density_results = calculate_density()

    print("Final Density Data:", density_results)
```

```python
def run_simulation():
    traci.start([sumo_binary, "-c", sumo_config])

    time = 0
    try:
        while time < 10800:
            traci.simulationStep()

        #차량 정보 가져오기
            if time % 100 == 0 :
                
                vehicle_ids = traci.vehicle. getIDList()
                for vehicle_id in vehicle_ids:
                    speed = traci.vehicle.getSpeed(vehicle_id) #차량 속도
                    position = traci.vehicle.getPosition(vehicle_id) #차량 위치
                    vehicle_type = traci.vehicle.getTypeID(vehicle_id)
                    print(f"Vehicle {vehicle_id} -> Speed: {speed:.2f}, Position: {position}, Type: {vehicle_type}")
        time+=1
    
    finally:
        traci.close()
        print("Simulation ended")
        
if __name__ == "__main__":
    run_simulation()
```
