```python
#분당 edge의 평균 속도 코드

start_sumo(sumo_binary, sumo_config)

time = 0
edge_id = "522941402"
times = []
speeds = []

minute_speeds = []
minute = 0

while time < 3600 :
    traci.simulationStep()

    minute_speeds.append(edge_speed_avg(edge_id))

    if time % 60 == 59:
        avg_speed = sum(minute_speeds) / len(minute_speeds)
        speeds.append(avg_speed)
        times.append(minute)
        minute += 1
        minute_speeds = []
        
    time += 1

plot_graph("km/h", times, speeds)
end_sumo()
```
