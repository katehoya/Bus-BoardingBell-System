```python
import traci

def start_sumo(sumo_binary, sumo_config):
    """
    sumo_binary: "sumo" 또는 "sumo-gui" 등 실행 파일
    sumo_config: -c 옵션에 넣을 .sumocfg 파일 경로
    """
    traci.start([sumo_binary, "-c", sumo_config])

def run_simulation(max_steps=1000):
    """
    시뮬레이션을 max_steps 동안 진행하면서
    각 스텝마다 차량 정보를 출력
    """
    for step in range(max_steps):
        # 시뮬레이션 1 스텝 진행
        traci.simulationStep()
        
        # 현재 시점에 존재하는 모든 차량 ID 가져오기
        veh_ids = traci.vehicle.getIDList()
        
        # 각 차량에 대해 정보 출력
        for vid in veh_ids:
            vtype = traci.vehicle.getTypeID(vid)       # 차량 타입
            speed = traci.vehicle.getSpeed(vid)        # 속도 (m/s)
            x, y  = traci.vehicle.getPosition(vid)     # 위치 (절대 좌표, m)
            # 필요하다면 차선 정보, 가속도, 경로 등 더 많이 가져올 수 있음
            
            print(f"[step={step}] ID={vid}, type={vtype}, speed={speed:.2f} m/s, pos=({x:.2f}, {y:.2f})")

    # 시뮬레이션 종료
    traci.close()

if __name__ == "__main__":
    # (1) sumo-gui로 실행하면 GUI가 뜨고
    # (2) sumo로 실행하면 CLI 모드 (화면 없음)
    SUMO_BINARY = "sumo-gui"       # 또는 "sumo"
    SUMO_CONFIG = "myConfig.sumocfg"

    start_sumo(SUMO_BINARY, SUMO_CONFIG)
    run_simulation(max_steps=300)  # 예: 300스텝(초) 진행
```
